```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('my graph')
plt.savefig('mine.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g*-')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave')
plt.plot(x,y,'r*')
```

```python
x=[2,8,10]
y=[11,16,9]
plt.bar(x,y)
plt.title('bar graph')
```

```python
a=np.array([23,45,6,7,76,32,56,76,45,67])
plt.hist(a)
plt.title('histogram')
```

BOXPLOT-MATPLOTLIB

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.title('boxplot')
```

Piechart

```python
labels='python','java','c++','ruby'
explode=(0.4,0,0,0)
colors=['gold','yellowgreen','lightcoral','lightskyblue']
sizes=[215,130,245,210]

plt.pie(sizes,labels=labels,explode=explode,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('piechart')
```
